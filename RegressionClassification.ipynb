## Introduction

In the supervised learning problem, there is set of data (the training data) that consists of a set of input data and its target data, which is the answer that the algorithm should produce. This is
usually written as a set of data $(\mathbf{x_i} , \mathbf{y_i} )$, where the inputs are $\mathbf{x_i}$ , the targets are $\mathbf{y_i}$ , and the $i$ index suggests that we have lots of pieces of data, indexed by $i$ running from 1 to some upper limit $N$. Note that the inputs and targets are written in boldface font to signify vectors, since each piece of data has values for several different features. The goal is to find the underlying law that relates data $\mathbf{x}$ to responses $\mathbf{y}$. In this pratical work, we will explore two important machine learning methods: The regression and the classification.

### The regression
Suppose that I gave you the following datapoints and asked you to tell me the value of the output when $x = 0.44$ (here, $x$, $y$, are not written in boldface font since they are scalars, as opposed to vectors). Since the value x = 0.44 isn’t in the examples given, you need to find some way to predict what value it has. You assume that the values come from some sort of function, and try to find out what the function is. Then you’ll be able to give the output value y for any given value of x. This is known as a regression problem in statistics: fit a mathematical function describing a curve, so that the curve passes as close as possible to all of the datapoints. It is generally a problem of function approximation or interpolation, working out the value between values that we know. So one thing that our machine learning algorithms can do is interpolate between data-points. This might not seem to be intelligent behaviour, or even very difficult in two dimensions, but it is rather harder in higher dimensional spaces.
Let us have a look at the simple problem of the regresion of a straight line with the linear model: $$\hat{y}=wx + b.$$ This 1D problem can be extended to a $n$ dimensional hyperplan: $$\hat{y} = \sum_{i=1}^n w_i x_i + b$$
We can also write this equation in the vector form: $$\hat{y}=\mathbf{W}^T \mathbf{X} + b \;\;\;(1)$$
The regression problem is to estimate the parameters $w_i, i=1,\dots,n$.

### The classification
The classification problem consists of taking input vectors and deciding which of $N$ classes they belong to, based on training from exemplars of each class. The most important point about the classification problem is that it is discrete — each example belongs to precisely one class, and the set of classes covers the whole possible output space.

## Model Training with loss function
For the regression problem, we expect to fit the mathematical model to the observation $y$.
we will minimize the square error:
$$J(W,b) = \frac{1}{2m} \sum_{i=1}^{m} (\hat{y}_i - y_i)^2$$
The classical method to minimize this error is to use the gradient descent method.
The linear model is function of $\mathbf{W}=(w_1, w_2,\cdots,w_n)$ and bias $b$. Thus $J(W,b)=J(\hat{y}_1(w_1,\cdots,w_n,b) ,\cdots,\hat{y}_m(w_1,\cdots,w_n,b)$. We can use the chaine rule to compute the gradient:
$$\frac{\partial{J}}{\partial{W}} = \sum_{i=1}^m \frac{\partial{J}}{\partial{\hat{y}_i}} \frac{\partial{\hat{y}_i}}{\partial{\mathbf{W}}} \;\;\;(2)$$
$$\frac{\partial{J}}{b} = \sum_{i=1}^m \frac{\partial{J}}{\partial{\hat{y}_i}} \frac{\partial{\hat{y}_i}}{b} \;\;\;(3)$$
The good approach is too start with the derivation of $J(\hat{y}_1, \cdots, \hat{y}_m)$ then with the derivation $\hat{y}_i(\mathbf{W},b)$. This is a first taste of the Back-propagation appoach for the gradient computation we will use in the next practical work.

Now, here are the gradients for our minimization problem:
$$\frac{\partial{J}}{\partial{w_k}} =  \frac{1}{m} \sum_{i=1}^{m} x_{i,k}(\hat{y}_i - y_i) \;\;\;(4)$$
$$\frac{\partial{J}}{\partial{b}} =  \frac{1}{m} \sum_{i=1}^{m}(\hat{y}_i - y_i) \;\;\;(5)$$

The first equation in matrix form is:

$$\frac{\partial{J}}{\partial{W}} =  \frac{1}{m} (\hat{Y} - Y).X^{T}$$

## Training the model
Finally, we can train the model for our problem.

The gradient descent update equation is as follows:

$$ W= W - \alpha \frac{\partial \mathcal{J} }{\partial W} \;\;\;(6)$$

$$ b = b - \alpha \frac{\partial \mathcal{J} }{\partial b} \;\;\;(7)$$

We will loop over the training set for a set number of iterations (feel free to tweak $\alpha$ and the number of iterations - these are hyperparameters).
We'll also periodically output the loss at that iteration.

```python
#Write the gradient method using equation (4) and (6)
def grads(X, Y, Y_pred):
    m = Y.shape[0]
    dW = (1/m) * np.dot((Y_pred - Y), X)
    db = (1/m) * np.sum(Y_pred - Y)
    return dW, db

#Write the descent gradient method
def gradient_descent(X, W, b, Y, Y_pred, alpha=1e-2):
  dW, db = grads(X, Y, Y_pred)
  W = W - alpha * dW
  b = b-alpha * db
  return W,b
```

Now we can train the model by calling the gradient_descent method with X_train, Y_train, W and b. The ouput of the method will be the trained weight to use in the model forward().

```python
#Your code to train your model

W = np.random.randn(X.shape[1])
b = np.random.randn(1)


for i in range(1000):
  y_pred_final = forward(X_train, W, b)
  W, b = gradient_descent(X_train, W, b, y_train, y_pred_final, alpha=1e-2)
  if i%100==0:
    loss= MSE_loss(y_train, y_pred_final)
    print("iteration:", i, "loss:", loss)
```

### The logistic log loss

The logistic loss function is $$ J(W,b) = \frac{-1}{m} \sum_{i=1}^{m} y^{(i)} \log(\hat{y}^{(i)}) + (1-y^{(i)}) \log(1-\hat{y}^{(i)})$$. Now we need to applied the same previous method by deriving the gradient of the loss function $J$. We need to compute $\frac{\partial{J}}{\partial{\hat{y}_i}}$, $\frac{\partial{\hat{y}_i}}{\partial{\mathbf{w_k}}}$ and $\frac{\partial{\hat{y}_i}}{\partial{b}}$:
* $\frac{\partial{J}}{\partial{\hat{y}_i}} = \frac{1}{m}(-\frac{y_i}{\hat{y}_i} + \frac{1-y_i}{(1-\hat{y}_i)})$
* $\frac{\partial{\hat{y}_i}}{\partial{w_k}} = x_{k,i}(\hat{y}_k(1-\hat{y}_k))$
* $\frac{\partial{\hat{y}_i}}{\partial{b}} = \hat{y}_k(1-\hat{y}_k)$

Then, $\frac{\partial{J}}{\partial{w_k}} = \frac{1}{m}\sum_{i=1}^m x_{i,k}(\hat{y}_i-y_i))$ and $\frac{\partial{J}}{\partial{b}} = \frac{1}{m}\sum_{i=1}^m (\hat{y}_i-y_i))$. 

Then $$\frac{\partial{J}}{\partial{W}} =  \frac{1}{m} (\hat{Y} - Y).X^{T}$$. We can notice that we get the same equation of gradient of the regression problem.


```python
#First: Weight and biais initialisation using np.random.randn(shape)

W = np.random.randn(X.shape[1])
b = np.random.randn(1)


#Then the equation (1)
def forward_logistic(X, W, b):
    z = np.dot(X, W) + b
    return 1 / (1 + np.exp(-z))

def logistic_loss(Y, Y_pred):
    loss = -np.mean(Y*np.log(Y_pred) + (1-Y)*np.log(1-Y_pred))
    return loss

#Finally the gradient descent method
def gradient_descent_log(X, W, b, Y, Y_pred, alpha=3e-2):
    m = len(Y)
    dW = (1/m) * np.dot(X.T, (Y_pred - Y))
    db = (1/m) * np.sum(Y_pred - Y)
    W = W - alpha * dW
    b = b - alpha * db
    return W, b
```

Now you can train your model! And test it

```python
#Your code to train your model and show it converges to 0

for i in range(1000):
    Y_pred = forward_logistic(X_train, W, b)
    loss = logistic_loss(y_train, Y_pred)
    W, b = gradient_descent_log(X_train, W, b, y_train, Y_pred, alpha=3e-2)
    if i % 100 == 0:
        print("iteration:", i, "loss:", loss)
```

For the method evaluation, you can use the F1 score metric. Please read [this](https://kobia.fr/classification-metrics-f1-score/) or [this](https://en.wikipedia.org/wiki/F-score)

```python
def F1_Score(Y_pred, Y):
    prediction = np.rint(Y_pred)
    truth_pos = (Y == np.ones_like(Y))
    truth_neg = (Y == np.zeros_like(Y))
    pred_pos = (prediction == np.ones_like(prediction))
    pred_neg = (prediction == np.zeros_like(prediction))

    true_pos = np.sum(np.logical_and(truth_pos,pred_pos))
    if true_pos == 0: #This prevents an undefined computation since precision=recall=0 
        return 0
    false_pos =np.sum(np.logical_and(truth_neg,pred_pos))
    false_neg =np.sum(np.logical_and(truth_pos,pred_neg))
    true_neg =np.sum(np.logical_and(truth_neg,pred_neg))

    precision = true_pos/(true_pos + false_pos)
    recall = (true_pos)/(true_pos + false_neg)
    F1_score = 2*(recall*precision) /(recall + precision)
    return F1_score 
```

```python
#Evolution in the test data
score_test = F1_Score(Y_pred_test, y_test)
print("Test score on test data", score_test)
```
